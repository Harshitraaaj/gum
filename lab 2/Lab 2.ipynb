```python
# Kindly use the Jupyter Notebook to run this program.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the built-in Iris dataset from seaborn
data = sns.load_dataset('iris')
print("âœ… Iris dataset loaded successfully.")
data.head()
```

```python
# Calculate Pearson correlation coefficient between sepal_length and petal_length
correlation = data[['sepal_length', 'petal_length']].corr(method='pearson')
print("ðŸ“ˆ Pearson Correlation Coefficient:\n")
print(correlation)
```

```python
# Compute the covariance matrix
covariance = data[['sepal_length', 'petal_length']].cov()
print("ðŸ“‰ Covariance Matrix:\n")
print(covariance)
```

```python
# Define x and y columns
x_col = 'sepal_length'
y_col = 'petal_length'

# Create scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(data[x_col], data[y_col], color='purple', alpha=0.6)
plt.xlabel(x_col.capitalize())
plt.ylabel(y_col.capitalize())
plt.title(f"Scatter Plot of {x_col} vs {y_col}")
plt.grid(True)
plt.show()
```

```python
# Select numeric columns (excluding species)
data_numeric = data.select_dtypes(include='number')

# Covariance matrix
cov_matrix = data_numeric.cov()
print("ðŸ“‰ Covariance Matrix (All Numeric Columns):\n")
print(cov_matrix)

# Correlation matrix
corr_matrix = data_numeric.corr()
print("\nðŸ“ˆ Correlation Matrix (All Numeric Columns):\n")
print(corr_matrix)
```

```python
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```
