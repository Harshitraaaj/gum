```python
# Kindly use the Jupyter Notebook to run this program.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load Iris dataset
iris = load_iris()
X = iris.data        # Feature matrix (4D)
y = iris.target      # Labels
target_names = iris.target_names

print("✅ Iris dataset loaded successfully.")
```

```python
# Standardize features to have mean=0 and variance=1
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# Reduce dimensions to 2 using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)

# Create DataFrame with PCA results
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df_pca['Species'] = y
```

```python
# Define species names and colors
species_labels = ['Setosa', 'Versicolor', 'Virginica']
colors = ['brown', 'hotpink', 'purple']

# Plot the PCA results
plt.figure(figsize=(7, 5))
for i, color in zip(np.unique(y), colors):
    plt.scatter(df_pca[df_pca['Species'] == i]['PC1'],
                df_pca[df_pca['Species'] == i]['PC2'],
                c=color, label=species_labels[i], edgecolor='k', alpha=0.7)

plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('🌸 PCA Visualization of Iris Dataset')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
